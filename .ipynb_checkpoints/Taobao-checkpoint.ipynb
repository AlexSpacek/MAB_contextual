```python
import pandas as pd, numpy as np, re
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.datasets import load_svmlight_file
from contextualbandits.online import BootstrappedUCB, EpsilonGreedy, BootstrappedTS, ExploreFirst
from sklearn.linear_model import LogisticRegression
from contextualbandits.online import EpsilonGreedy, LogisticTS, ParametricTS, LinUCB, LinTS
from contextualbandits.evaluation import evaluateRejectionSampling

dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
y = dataset.values
features = np.ascontiguousarray(np.ones((y.shape[0],1)))
```

```python
nchoices = y.shape[1]
beta_prior = ((3./nchoices, 4), 2) # until there are at least 2 observations of each class, will use this prior
beta_prior_ucb = ((5./nchoices, 4), 2)
np.random.seed(456)
actions_random = np.random.randint(nchoices, size = y.shape[0]) #this simulates a dataset which is not fully labeled
rewards_actions = y[np.arange(y.shape[0]), actions_random] # so this would be 0/1 from ad history

online_policy1 = BootstrappedUCB(LogisticRegression(C=0.1, solver="lbfgs", max_iter=15000),
                                y.shape[1], random_state=1111, beta_prior = beta_prior)
online_policy2 = ExploreFirst(LogisticRegression(C=0.1, solver="lbfgs", max_iter=15000),
                                y.shape[1], explore_rounds=5000,beta_prior = None, random_state=2222)
online_policy3 = EpsilonGreedy(LogisticRegression(C=0.1, solver="lbfgs", max_iter=15000),
                                y.shape[1], explore_prob=0.2, decay=0.9999, random_state=3333,beta_prior = beta_prior)
online_policy4 = LinUCB(y.shape[1], random_state=3333, beta_prior = beta_prior)

online_policy5 = LinTS(y.shape[1], random_state=3333, beta_prior = beta_prior)

online_policy6 = BootstrappedTS(LogisticRegression(C=0.1, solver="lbfgs", max_iter=15000),
                                y.shape[1], random_state=3333,beta_prior = beta_prior)

```

```python
[cum_r1, cum_n1, cum_r_list1, cum_mean_r_list1] = evaluateRejectionSampling(online_policy1,
                          X = features,
                          a = actions_random,
                          r = rewards_actions,
                          online = True,
                          start_point_online = 0,
                          batch_size = 10,
                          random_state = 5678)
[cum_r2, cum_n2, cum_r_list2, cum_mean_r_list2] = evaluateRejectionSampling(online_policy2,
                          X = features,
                          a = actions_random,
                          r = rewards_actions,
                          online = True,
                          start_point_online = 0,
                          batch_size = 10,
                          random_state = 5678)
[cum_r3, cum_n3, cum_r_list3, cum_mean_r_list3] = evaluateRejectionSampling(online_policy3,
                          X = features,
                          a = actions_random,
                          r = rewards_actions,
                          online = True,
                          start_point_online = 0,
                          batch_size = 10,
                          random_state = 5678)
[cum_r4, cum_n4, cum_r_list4, cum_mean_r_list4] = evaluateRejectionSampling(online_policy3,
                          X = features,
                          a = actions_random,
                          r = rewards_actions,
                          online = True,
                          start_point_online = 0,
                          batch_size = 10,
                          random_state = 5678)
[cum_r5, cum_n5, cum_r_list5, cum_mean_r_list5] = evaluateRejectionSampling(online_policy3,
                          X = features,
                          a = actions_random,
                          r = rewards_actions,
                          online = True,
                          start_point_online = 0,
                          batch_size = 10,
                          random_state = 5678)
[cum_r6, cum_n6, cum_r_list6, cum_mean_r_list6] = evaluateRejectionSampling(online_policy3,
                          X = features,
                          a = actions_random,
                          r = rewards_actions,
                          online = True,
                          start_point_online = 0,
                          batch_size = 10,
                          random_state = 5678)
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
%matplotlib inline

rcParams['figure.figsize'] = 25, 10
lwd = 5
cmap = plt.get_cmap('tab20')
colors=plt.cm.tab20(np.linspace(0, 1, 20))

ax = plt.subplot(111)
plt.plot(cum_mean_r_list1, label="BootstrappedUCB",linewidth=lwd,color=colors[0])
plt.plot(cum_mean_r_list2, label="ExploreFirst (50%)",linewidth=lwd,color=colors[2])
plt.plot(cum_mean_r_list3, label="Epsilon Greedy (e=0.2)",linewidth=lwd,color=colors[4])
plt.plot(cum_mean_r_list4, label="LinUCB",linewidth=lwd,color=colors[6])
plt.plot(cum_mean_r_list5, label="LinTS",linewidth=lwd,color=colors[8])
plt.plot(cum_mean_r_list6, label="BootstrappedTS",linewidth=lwd,color=colors[10])

# import warnings
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 1.25])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 0.2),
          fancybox=True, ncol=2, prop={'size':20})


plt.tick_params(axis='both', which='major', labelsize=25)
# plt.xticks([i*20 for i in range(8)], [i*1000 for i in range(8)])


plt.xlabel('Rounds ', size=30)
plt.ylabel('Cumulative Mean Reward', size=30)
plt.title('Comparison of Online Contextual Bandit Policies\n(Base Algorithm is Logistic Regression)\n\nSmall Ad Dataset\n(10 categories, 0 attributes)',size=30)
plt.grid()
plt.show()
```
