```python
import pandas as pd, numpy as np, re
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.datasets import load_svmlight_file
from contextualbandits.online import BootstrappedUCB, EpsilonGreedy, BootstrappedTS, ExploreFirst
from sklearn.linear_model import LogisticRegression
from contextualbandits.online import EpsilonGreedy, LogisticTS, ParametricTS, LinUCB, LinTS
from contextualbandits.evaluation import evaluateRejectionSampling

dataset = pd.read_csv('Taobao\\train.csv')
# y = dataset.values
# features = np.ascontiguousarray(np.ones((y.shape[0],1)))
```

```python
take_first_N_values = 1000000
dataset = dataset.head(take_first_N_values)
```

```python
dataset = dataset[dataset['cate_id'].map(dataset['cate_id'].value_counts()) > take_first_N_values/100]
dataset.head()
```

```python
rewards = dataset['clk']
actions = dataset['cate_id']
```

```python
print(rewards.shape,actions.shape)
```

```python
'taking ad category column and transforming it into NxK matrix where K is the number of categories'
from sklearn.preprocessing import LabelEncoder, LabelBinarizer, OneHotEncoder
le_act = LabelEncoder()
actions = le_act.fit_transform(actions)
print(actions.shape)
```

```python
dataset = dataset[['final_gender_code','age_level','shopping_level','occupation','pvalue_level']]
for column in dataset:
    display(dataset[column].value_counts(dropna=False))
```

```python
'replace nan values with random value, but keeping the same statistics'
import random
for col in dataset:
    data = dataset[col]
    mask = data.isnull()
    samples = random.choices( data[~mask].values , k = mask.sum() )
    dataset[col].iloc[mask] = samples
for column in dataset:
    display(dataset[column].value_counts(dropna=False))   
```

```python
'encoding features using one hot encoding'
## columns_to_encode = ['final_gender_code','age_level','shopping_level','occupation','pvalue_level']
# enc = OneHotEncoder()
# features = enc.fit_transform(dataset.astype(str)).toarray()
# print(features.shape)
features = dataset.values
```

```python
# feature_names = enc.get_feature_names_out()
# print(feature_names)
# print(features.shape)
# columns_to_delete = [np.where(feature_names == 'final_gender_code_nan')[0][0],
#                      np.where(feature_names == 'age_level_nan')[0][0],
#                      np.where(feature_names == 'shopping_level_nan')[0][0],
#                      np.where(feature_names == 'occupation_nan')[0][0],
#                      np.where(feature_names == 'pvalue_level_nan')[0][0]]
# features = np.delete(features, columns_to_delete, 1)
# print(features.shape)
```

```python
# print(features[0:5,:])
```

```python
nchoices = len(np.unique(actions))
print(nchoices)
```

```python
beta_prior = ((3./nchoices, 4), 2) # until there are at least 2 observations of each class, will use this prior
beta_prior_ucb = ((5./nchoices, 4), 2)
np.random.seed(456)
# actions_random = np.random.randint(nchoices, size = y.shape[0]) #this simulates a dataset which is not fully labeled
# rewards_actions = y[np.arange(y.shape[0]), actions_random] # so this would be 0/1 from ad history

BootstrappedUCB_policy = BootstrappedUCB(LogisticRegression(C=1, solver="lbfgs", max_iter=15000),
                                nchoices, random_state=1111)
BootstrappedTS_policy = BootstrappedTS(LogisticRegression(C=1, solver="lbfgs", max_iter=15000),
                                nchoices, random_state=1111)
EpsilonGreedy_policy = EpsilonGreedy(LogisticRegression(C=1, solver="lbfgs", max_iter=15000),
                                nchoices, explore_prob=0.2, decay=0.9999, random_state=2222)
Random_policy = EpsilonGreedy(LogisticRegression(C=1, solver="lbfgs", max_iter=15000),
                                nchoices, explore_prob=0.9999999, decay=0.9999999, random_state=3333)
LogisticTS_policy = LogisticTS(nchoices,  random_state=4444)
# online_policy6 = BootstrappedTS(LogisticRegression(C=0.1, solver="lbfgs", max_iter=15000),
#                                 y.shape[1], random_state=3333,beta_prior = beta_prior)

```

```python
[cum_r1, cum_n1, cum_r_list1, cum_mean_r_list1] = evaluateRejectionSampling(BootstrappedUCB_policy,
                          X = features,
                          a = actions,
                          r = rewards,
                          online = True,
                          start_point_online = 0,
                          batch_size = 1000)
```

```python
[cum_r2, cum_n2, cum_r_list2, cum_mean_r_list2] = evaluateRejectionSampling(EpsilonGreedy_policy,
                          X = features,
                          a = actions,
                          r = rewards,
                          online = True,
                          start_point_online = 0,
                          batch_size = 1000)
```

```python
[cum_r3, cum_n3, cum_r_list3, cum_mean_r_list3] = evaluateRejectionSampling(Random_policy,
                          X = features,
                          a = actions,
                          r = rewards,
                          online = True,
                          start_point_online = 0,
                          batch_size = 1000)
```

```python
[cum_r4, cum_n4, cum_r_list4, cum_mean_r_list4] = evaluateRejectionSampling(LogisticTS_policy,
                          X = features,
                          a = actions,
                          r = rewards,
                          online = True,
                          start_point_online = 0,
                          batch_size = 1000)
```

```python
[cum_r5, cum_n5, cum_r_list5, cum_mean_r_list5] = evaluateRejectionSampling(BootstrappedTS_policy,
                          X = features,
                          a = actions,
                          r = rewards,
                          online = True,
                          start_point_online = 0,
                          batch_size = 1000)
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
%matplotlib inline

rcParams['figure.figsize'] = 25, 10
lwd = 5
cmap = plt.get_cmap('tab20')
colors=plt.cm.tab20(np.linspace(0, 1, 20))

ax = plt.subplot(111)
plt.plot(cum_mean_r_list1, label="BootstrappedUCB",linewidth=lwd,color=colors[0])
plt.plot(cum_mean_r_list2, label="Epsilon Greedy (e=0.2)",linewidth=lwd,color=colors[2])
plt.plot(cum_mean_r_list3, label="Random sampling",linewidth=lwd,color=colors[4])
plt.plot(cum_mean_r_list4, label="LogisticTS",linewidth=lwd,color=colors[6])
plt.plot(cum_mean_r_list5, label="BootstrappedUCBS",linewidth=lwd,color=colors[8])
plt.plot(np.repeat(np.sum(rewards)/len(rewards),len(cum_mean_r_list1)), label="Overall CTR",linewidth=lwd,color=colors[1],ls='dashed')

# import warnings
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 1.25])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 0.2),
          fancybox=True, ncol=2, prop={'size':20})


plt.tick_params(axis='both', which='major', labelsize=25)
# plt.xticks([i*20 for i in range(8)], [i*1000 for i in range(8)])


plt.xlabel('Rounds ', size=30)
plt.ylabel('Cumulative Mean CTR', size=30)
plt.title('Comparison of Online Contextual Bandit Policies\n(Logistic regression updated each 1000 samples)\n\nTaobao Dataset (first 5M samples)',size=30)
plt.grid()
plt.ylim((0.04,0.07))
plt.show()
```

```python
'we can see that the random sampling produced similar results to the average value, hence the dataset policy was not biased and we can used the evaluator'
```
